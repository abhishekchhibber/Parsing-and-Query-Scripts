```python
def get_links(page_url):
    opener_in = urllib.request.build_opener()
    opener_in.addheaders = [('User-agent', "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:36.0) Gecko/20100101 Firefox/36.0")]
    html_in = opener_in.open(page_url, timeout=25)
    time.sleep(25)
    html_read_in = html_in.read()
    soup_in = bs(html_read_in, 'html.parser')
    items_in  = soup_in.findAll("div",{"class":"name-info"})
    for item in items_in:
        print (item.a['href'])
#         global link_list
        link_list.append(item.a['href'])
    
        
    return
```
